```python
import tkinter as tk
from tkinter import messagebox

# Question data
quiz_data = [
    {"question": "What is the capital of France?", "options": ["Paris", "London", "Berlin", "Madrid"], "answer": "Paris"},
    {"question": "What is the square root of 64?", "options": ["6", "8", "10", "12"], "answer": "8"},
    {"question": "Which language is known as the 'Mother of all Languages'?", "options": ["Sanskrit", "Latin", "Greek", "Arabic"], "answer": "Sanskrit"},
    {"question": "Who wrote 'To Kill a Mockingbird'?", "options": ["Harper Lee", "J.K. Rowling", "Ernest Hemingway", "Mark Twain"], "answer": "Harper Lee"},
]

# Initialize score
score = 0
current_question = 0

def display_question():
    """Update the question and options on the screen."""
    question_label.config(text=quiz_data[current_question]["question"])
    for i, option in enumerate(quiz_data[current_question]["options"]):
        options_buttons[i].config(text=option, value=option)

def next_question():
    """Proceed to the next question or show the final score."""
    global current_question, score
    selected = answer_var.get()
    if selected == quiz_data[current_question]["answer"]:
        score += 1
    current_question += 1

    if current_question < len(quiz_data):
        display_question()
    else:
        messagebox.showinfo("Quiz Complete", f"Your score is {score}/{len(quiz_data)}!")
        root.destroy()

# Create the main window
root = tk.Tk()
root.title("Python Quiz App")

# Question label
question_label = tk.Label(root, text="", font=("Arial", 16), wraplength=400, justify="center")
question_label.pack(pady=20)

# Variable to store selected answer
answer_var = tk.StringVar()

# Option buttons
options_buttons = []
for _ in range(4):
    btn = tk.Radiobutton(root, text="", variable=answer_var, font=("Arial", 14), value="", indicatoron=False, width=20)
    btn.pack(pady=5)
    options_buttons.append(btn)

# Next button
next_button = tk.Button(root, text="Next", command=next_question, font=("Arial", 14))
next_button.pack(pady=20)

# Start with the first question
display_question()

# Run the main loop
root.mainloop()
```
